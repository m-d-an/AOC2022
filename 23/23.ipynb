```python
from operator import itemgetter
```

```python
demo = False
if (demo):
    file = 'input_ex1.txt'
else:
    # Puzzle input
    file = 'input.txt'
```

```python
file = open(file, 'r')
input = file.readlines()
file.close()
```

```python
def rect_points():
    max_i = max(elves,key=itemgetter(0))[0]
    max_j = max(elves,key=itemgetter(1))[1]
    min_i = min(elves,key=itemgetter(0))[0]
    min_j = min(elves,key=itemgetter(1))[1]  
    return min_i, max_i, min_j, max_j
```

```python
def print_elves():
    min_i, max_i, min_j, max_j = rect_points()
    for i in range(min_i, max_i+1, 1):
        line = ""
        for j in range(min_j, max_j+1, 1):
            if (i,j) in elves:
                line += "#"
            else: 
                line += "."
        print(line)
    return
```

```python
elves = []
for i,line in enumerate(input):
    for j,c in enumerate(line.strip()):
        if c == '#':
            elves.append((i,j))
```

```python
def has_neighbour(elf):
    for i in range(elf[0]-1,elf[0]+2,1):
        for j in range(elf[1]-1,elf[1]+2,1):
            if (i,j) != elf and (i,j) in elves:
                return True
    return False
```

```python
def get_movement(action, elf, counter):
    if counter == 5:
        return None
    if action == 0:
        # North
        if (elf[0]-1,elf[1]-1) in elves or (elf[0]-1,elf[1]) in elves or (elf[0]-1,elf[1]+1) in elves:
            return get_movement(1, elf, counter+1)
        else:
            return (elf[0]-1,elf[1])
    if action == 1:
        # South
        if (elf[0]+1,elf[1]-1) in elves or (elf[0]+1,elf[1]) in elves or (elf[0]+1,elf[1]+1) in elves:
            return get_movement(2, elf, counter+1)
        else:
            return (elf[0]+1,elf[1])
    if action == 2:
        # West
        if (elf[0]-1,elf[1]-1) in elves or (elf[0],elf[1]-1) in elves or (elf[0]+1,elf[1]-1) in elves:
            return get_movement(3, elf, counter+1)
        else:
            return (elf[0],elf[1]-1)
    if action == 3:
        # East
        if (elf[0]-1,elf[1]+1) in elves or (elf[0],elf[1]+1) in elves or (elf[0]+1,elf[1]+1) in elves:
            return get_movement(0, elf, counter+1)
        else:
            return (elf[0],elf[1]+1)
```

```python
print("initial state:")
print_elves()
action_counter = 0

# going through rounds
for i in range(10):  
    elf_movements = {}
    duplicate_movements = []
    # movement propositions
    for elf in elves:
        if has_neighbour(elf):
            movement = get_movement(action_counter, elf, 1)
            if movement != None:
                if movement in elf_movements.values():
                    duplicate_movements.append(movement)
                else:
                    elf_movements[elf] = movement
    # movements
    for k,v in elf_movements.items():
        # ignore duplicate movement
        if v not in duplicate_movements:
            elves.remove(k)
            elves.append(v)
    # end of round
    action_counter += 1
    if action_counter == 4:
        action_counter = 0
    print("round "+str(i+1))
    print_elves()
```

```python
def count_empty_ground():
    min_i, max_i, min_j, max_j = rect_points()
    empty_ground = 0
    for i in range(min_i, max_i+1, 1):
        for j in range(min_j, max_j+1, 1):
            if (i,j) in elves:
                continue
            else: 
                empty_ground += 1
    return empty_ground
```
