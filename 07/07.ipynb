```python
from anytree import Node, RenderTree, NodeMixin
```

```python
file = open('input.txt', 'r')
input = file.readlines()
file.close()
```

```python
class Dir(NodeMixin):
    def __init__(self, name, parent=None) -> None:
        super(Dir, self).__init__()
        self.name = name
        self.size = 0
        self.parent = parent

    def add_size(self, size):
        self.size += size
        if (self.parent != None):
            self.parent.add_size(size)
```

```python
class File(NodeMixin):
    def __init__(self, name, size, parent) -> None:
        super(File, self).__init__()
        self.name = name
        self.size = size
        self.parent = parent
        self.parent.add_size(size)
```

```python
def print_tree(root):
    for pre, fill, node in RenderTree(root):
        treestr = u"%s%s" % (pre, node.name)
        print(treestr.ljust(8), node.size)
```

```python
def solution_a(root):
    sum = 0
    for pre, fill, node in RenderTree(root):
        if isinstance(node, Dir) and node.size<=100000:
            sum += node.size
    print(sum)
```

```python
def solution_b(root):
    total_space = 70000000
    free_space = total_space - root.size
    missing_space = 30000000 - free_space

    chosen_dir = root
    for pre, fill, node in RenderTree(root):
        if isinstance(node, Dir) and node.size>=missing_space and node.size<chosen_dir.size:
            chosen_dir = node
    print(chosen_dir.size)
```

```python
active_dir = None
for line in input:
    if (line.startswith('$ cd /')):
        root_dir = Dir(line.split()[2])
        active_dir = root_dir
    elif (line.startswith('$ cd ..')):
        active_dir = active_dir.parent
    elif (line.startswith('$ cd ')):
        active_dir = Dir(line.split()[2], active_dir)
    elif (line.startswith('$ ls')):
        continue
    elif (line.startswith('dir')):
        continue
    else:
        split = line.split()
        File(split[1], int(split[0]), active_dir)
```

```python
print_tree(root_dir)
```

```python
solution_a(root_dir)
```

```python
solution_b(root_dir)
```
