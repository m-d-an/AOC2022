```python
import numpy as np
```

```python
example = False
if (example):
    # Example 1
    file = 'input_ex1.txt'
else:
    # Puzzle input
    file = 'input.txt'
```

```python
file = open(file, 'r')
input = file.readlines()
file.close()
```

```python
def print_grid(grid):
    for i in range(len(grid)):
        line = ""
        for j in range(len(grid[i])):
            if(grid[i,j] == 0):
                line += '.'
            if(grid[i,j] == 1):
                line += '#'
            if(grid[i,j] == 5):
                line += '+'
            if(grid[i,j] == 2):
                line += 'o'
        print(line)
```

```python
# parse input data
rock_lines_v = []
rock_lines_h = []
for line in input:
    v = []
    h = []
    for s in line.split():
        coords = s.split(',')
        if(coords[0].isnumeric()):
            v.append(int(coords[0]))
            h.append(int(coords[1]))
    rock_lines_v.append(v)
    rock_lines_h.append(h)
```

```python
# add floor input
if (example):
    rock_lines_v.append([480,520])
else:
    rock_lines_v.append([333,667])
rock_lines_h.append([max(map(max, rock_lines_h))+2,max(map(max, rock_lines_h))+2])
```

```python
# create grid
sand_entry_point = (0, 500)
v_coords = (min(map(min, rock_lines_v)),max(map(max, rock_lines_v)))
h_coords = (min(map(min, rock_lines_h)),max(map(max, rock_lines_h)))
v_len = v_coords[1]-v_coords[0]+1
h_len = h_coords[1]-sand_entry_point[0]+1
grid = np.zeros((h_len, v_len))
```

```python
# sand entry point
grid[sand_entry_point[0], sand_entry_point[1]-v_coords[0]] = 5
```

```python
# create rock lines
for i in range(len(rock_lines_h)):
    s_r = s_c = j = 0
    for row,col in zip(rock_lines_h[i], rock_lines_v[i]):
        if j>0:
            if s_r > row:
                grid[row:s_r+1, col-v_coords[0]] = 1
            else:
                grid[s_r:row+1, col-v_coords[0]] = 1
            if s_c > col-v_coords[0]:
                grid[row, col-v_coords[0]:s_c+1] = 1
            else:
                grid[row, s_c:col-v_coords[0]+1] = 1
        j += 1
        s_r = row
        s_c = col-v_coords[0]
```

```python
import time
from IPython.display import clear_output

def drop_sand(point, print= False):
    # falling sand
    current_point = (point[0],point[1]-v_coords[0])
    is_moving = True
    while(is_moving):
        if current_point[0]+1 == len(grid):
            grid[current_point[0],current_point[1]] = 0
            return False     
        elif grid[current_point[0]+1,current_point[1]] == 0:
            grid[current_point[0],current_point[1]] = 0
            current_point = (current_point[0]+1,current_point[1])
            grid[current_point[0],current_point[1]] = 2
        elif current_point[1]-1 < 0:
            grid[current_point[0],current_point[1]] = 0
            return False
        elif grid[current_point[0]+1,current_point[1]-1] == 0:  
            grid[current_point[0],current_point[1]] = 0
            current_point = (current_point[0]+1,current_point[1]-1)
            grid[current_point[0],current_point[1]] = 2
        elif current_point[1]+1 == len(grid[0]):
            grid[current_point[0],current_point[1]] = 0
            return False
        elif grid[current_point[0]+1,current_point[1]+1] == 0:  
            grid[current_point[0],current_point[1]] = 0
            current_point = (current_point[0]+1,current_point[1]+1)
            grid[current_point[0],current_point[1]] = 2
        elif current_point == (point[0],point[1]-v_coords[0]):
            grid[current_point[0],current_point[1]] = 2
            return False
        else:
            is_moving = False

        # reset sand entry point
        grid[point[0],point[1]-v_coords[0]] = 5

        if (print):
            clear_output(wait=True)
            print_grid(grid)
            time.sleep(0.05)
    return True
```

```python
# solution b
sand_units = 1
while(drop_sand(sand_entry_point, False)):
    sand_units += 1
clear_output(wait=True)
print_grid(grid)
print(sand_units)
```
